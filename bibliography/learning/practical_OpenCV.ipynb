```python
import numpy as np
import cv2 as cv 
import scipy as sp
import matplotlib.pyplot as plt
```

### Loading, displaying and Saving



```python
def show_image(image, name="Image"):
     image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
     plt.imshow(image_rgb)
     plt.axis("off")
     plt.show()
```

```python
lena_path = "../../images/lena_color.tiff"
lena = cv.imread(lena_path)

show_image(lena, "Lena")
```

### Image basics

```python
def display_images(images: list, titles: list = None, cmap: str = None):
    fig, axs = plt.subplots(1, len(images), figsize=(15, 5))
    
    if len(images) == 1:
        axs = [axs]

    for i, img in enumerate(images):
        axs[i].imshow(img, cmap=cmap)
        axs[i].axis("off")
        
        if titles and i < len(titles):
            axs[i].set_title(titles[i])

    plt.show()
```

```python
lena.shape

blue = lena[:, :, 0]  # blue
green = lena[:, :, 1]  # green
red = lena[:, :, 2]  # red

type(blue)
```

```python
zeros_arr = np.zeros_like(blue)

lst = [np.stack([blue, zeros_arr, zeros_arr], axis=-1),
       np.stack([zeros_arr, green, zeros_arr], axis=-1), 
       np.stack([zeros_arr, zeros_arr, red], axis=-1)]
display_images(lst)
```
